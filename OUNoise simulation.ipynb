# OUNoise Simulation

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from utils.ounoise import OUNoise
from collections import deque
```

```python
#Parameters
exploration_mu = 0
exploration_theta = 0.1
exploration_sigma = 3.5
ounoise = OUNoise(size = 1, mu= exploration_mu, theta=exploration_theta, sigma=exploration_sigma)
window = 500
```

```python
#Initializations
avg_noise = []
for i in range(500):
    avg_noise.append(np.NaN)

noise_samples = []
noise_samples_mean = deque(maxlen=window)
```

```python
#Generating the noise
for i in range(1000):
    noise_state = ounoise.sample()
    noise_samples_mean.append(noise_state)
    noise_samples.append(noise_state)
    if i >= window:
        avg_noise_sample = np.mean(noise_samples_mean)
        avg_noise.append(avg_noise_sample)
```

```python
fig = plt.figure(figsize=(15,7))
fig.suptitle("Noise evolution")
plt.plot(noise_samples, color = 'grey', alpha = 0.4)
plt.plot(avg_noise)
plt.legend(['Noise state', 'Average Noise'], fontsize = 'large')
```
